```python
# Libreries for linear regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Datetime libreries
import datetime as dt
```

```python
# Converting date to ordinal
```

```python

ind['date'] = pd.to_datetime(ind['date'])
```

```python
ind['date']=ind['date'].map(dt.datetime.toordinal)
```

```python
ind.head()
```

```python
# Fitting model for linear regression
```

```python
x=ind["date"]
y=ind["total_cases"]
```

```python
# Spliting the whole data for train and test the model (70% for train and 30% for test)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
# Fittting of Linear Regression # Prediction # Accuracy Score
```

```python
l=LinearRegression()
```

```python
#l.fit(x_train,y_train)
l.fit(np.array(x_train).reshape(-1,1),np.array(y_train).reshape(-1,1))
```

```python
y_pred=l.predict(np.array(x_test).reshape(-1,1))
y_pred
```

```python
l.score(np.array(x_test).reshape(-1,1),np.array(y_test).reshape(-1,1))
```
